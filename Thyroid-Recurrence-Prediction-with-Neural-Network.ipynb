```python
df = pd.read_csv("/kaggle/input/thyroid-disease-data/Thyroid_Diff.csv")
df.head()
```

```python
df.shape
```

```python
X = df.drop(['Recurred'], axis = 1)
X.head()
```

```python

from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
```

```python
def one_hot_encode(df):
    categorical_cols = df.select_dtypes(include=['object', 'category']).columns
    ohe = OneHotEncoder(sparse=False, drop='first')
    ohe_df = pd.DataFrame(ohe.fit_transform(df[categorical_cols]), columns=ohe.get_feature_names_out(categorical_cols))
    df = df.drop(categorical_cols, axis=1)
    df = pd.concat([df, ohe_df], axis=1)
    return df

def min_max_scale_column(df, column_name):
    scaler = MinMaxScaler()
    df[column_name] = scaler.fit_transform(df[[column_name]])
    return df

X = one_hot_encode(X)
```

```python
X.head()
```

```python
X = min_max_scale_column(X, 'Age')
X.head()
```

```python
X.shape
```

```python
y = df['Recurred']
y = y.map({'Yes': 1, 'No': 0})
y.head()
```

```python
y.nunique()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np

class ThyroidModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(ThyroidModel, self).__init__()
        self.hidden = nn.Linear(input_size, hidden_size)
        self.output = nn.Linear(hidden_size, output_size)
        self.activation = nn.ReLU()
        
    def forward(self, x):
        x = self.activation(self.hidden(x))
        x = self.output(x)
        return x
    
    def fit(self, x_train, y_train, epochs=10, learning_rate=0.001):
        criterion = nn.BCEWithLogitsLoss() 
        optimizer = optim.Adam(self.parameters(), lr=learning_rate)
        

        if isinstance(x_train, pd.DataFrame):
            x_train = torch.tensor(x_train.values, dtype=torch.float32)
        elif isinstance(x_train, np.ndarray):
            x_train = torch.tensor(x_train, dtype=torch.float32)
        
        if isinstance(y_train, pd.Series):
            y_train = torch.tensor(y_train.values, dtype=torch.float32)
        elif isinstance(y_train, np.ndarray):
            y_train = torch.tensor(y_train, dtype=torch.float32)
        
        for epoch in range(epochs):
            self.train()
            optimizer.zero_grad()
            outputs = self(x_train).squeeze()
            if outputs.dim() == 0:
                outputs = outputs.unsqueeze(0)
            loss = criterion(outputs, y_train)
            loss.backward()
            optimizer.step()
            print(f'Epoch {epoch+1}/{epochs}, Loss: {loss.item()}')
    
    def predict(self, x):
        self.eval()
        if isinstance(x, pd.DataFrame):
            x = torch.tensor(x.values, dtype=torch.float32)
        elif isinstance(x, np.ndarray):
            x = torch.tensor(x, dtype=torch.float32)
        
        with torch.no_grad():
            outputs = self(x).squeeze()
        return torch.sigmoid(outputs)  # Assuming binary classification
    
    def score(self, x, y_true):
        y_pred = self.predict(x)
        y_pred = torch.round(y_pred)
        if isinstance(y_true, pd.DataFrame) or isinstance(y_true, pd.Series):
            y_true = torch.tensor(y_true.values, dtype=torch.float32)
        elif isinstance(y_true, np.ndarray):
            y_true = torch.tensor(y_true, dtype=torch.float32)
        accuracy = accuracy_score(y_true.numpy(), y_pred.numpy())
        return accuracy
```

```python
input_size = 40
hidden_size = 20
output_size = 1

model = ThyroidModel(input_size, hidden_size, output_size)
model.fit(x_train, y_train, epochs=500)

# Predict and score
predictions = model.predict(x_test)
accuracy = model.score(x_test, y_test)

print(f'Accuracy: {accuracy}')
```
